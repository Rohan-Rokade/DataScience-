#  BAR CHART 

# Importing matplotlib

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

###  plt.bar()  
Used to create a bar plot (vertically )

```python
import pandas as pd
import numpy as np

A=['a','b','c','a','a','b','a']
B=[10,20,10,20,30,60,5]
data={'A1':A,'B1':B}
df=pd.DataFrame(data)

print(df.head())

pitem=df.groupby('A1').sum()
print(pitem)

x=pitem.index.tolist()
y=np.array(pitem.values.tolist()).flatten()
plt.bar(x,y)
```

### Color :
The keyword argument color to set the color of the bars:

```python
plt.bar(x, y, color = "red") 
plt.show()
```

```python
plt.bar(x, y, color = ["red","purple","black"]) 
plt.show()
```

## Bar Width
The bar() takes the keyword argument width to set the width of the bars
width = 0.2

```python
plt.bar(x, y, color = "r",width=0.2) 
plt.show()
```

## plt.grid()

If you want grid lines, you can do that. Add the function call .grid() with color, linestyle, width and axis. 
Optionally you can add an alpha value.e.g alpha=0.7

```python
plt.bar(x, y, color = "r",width=0.2) 
plt.grid(color='purple', linestyle='--', linewidth=2, axis='y')
plt.show()
```

## **Using xlabel,ylabel , title and legend**

### **plt.xlabel()** 
- Label for x-axis

### **plt.ylabel()**
- Label for y axis 

### **plt.legend()**
The Legend class is a container of legend handles and legend texts.

### plt.title()
parameters - And we can set irt as,
fontsize=18

```python
plt.bar(x, y, color = "r",width=0.2) 
plt.xlabel('Class',fontsize=16)
plt.ylabel('Sum ',fontsize=16)
plt.title('Distribution of class',fontsize=20)
plt.grid(color='purple', linestyle='--', linewidth=2, axis='y')
plt.show()
```

# Bar Graph horizontally

## plt.barh()
To plot bar chart horizontally 

You have to use height parameter instead of width for plt.barh()

```python
plt.barh(x, y, color = "m",height=0.2) 
plt.xlabel('Class')
plt.ylabel('Sum ')
plt.title('Distribution of class')
plt.grid(color='black', linestyle='--', linewidth=2, axis='y')
plt.show()
```

# Multiple  bar charts
You can plot multiple bar charts in one plot.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

data1 = [23,85, 72, 43, 52]
data2 = [42, 35, 21, 16, 9]

w=0.3
plt.bar(np.arange(len(data1)), data1, width=w,color='r')
plt.bar(np.arange(len(data2))+w, data2, width=w,color='k')
plt.show()
```

# Stack charts
You can stack bar charts on top of each other. That is particulary useful when you multiple values combine into something greater.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt

data1 = [23,85, 72, 43, 52]
data2 = [42, 35, 21, 16, 9]
plt.bar(range(len(data1)), data1,color='red')
plt.bar(range(len(data2)), data2, bottom=data1,color='blue')
plt.show()
```

```python
countries = ['USA', 'GB', 'China', 'Russia', 'Germany']
bronzes = np.array([38, 17, 26, 19, 15])
silvers = np.array([37, 23, 18, 18, 10])
golds = np.array([46, 27, 26, 19, 17])
ind = [x for x, _ in enumerate(countries)]

plt.bar(ind, golds, width=0.8, label='golds', color='gold', bottom=silvers+bronzes)
plt.bar(ind, silvers, width=0.8, label='silvers', color='silver', bottom=bronzes)
plt.bar(ind, bronzes, width=0.8, label='bronzes', color='#CD853F')

plt.xticks(ind, countries)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.legend(loc="upper right")
plt.title("2012 Olympics Top Scorers")

plt.show()
```
