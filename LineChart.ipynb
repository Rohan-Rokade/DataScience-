# Line Chart using Matplotlib

##  Importing matplotlib

```python
import matplotlib.pyplot as plt
```

## **Plotting a simple line chart** 

### **plt.plot()** 
creates a plot

```python
h=[150,160,165,185]
w=[70,80,90,100]
#draw the plott                       
plt.plot(h,w)
plt.savefig('Test.png')
```

## **Using labels and legends**

### **plt.xlabel()** 
- Label for x-axis

### **plt.ylabel()**
- Label for y axis 

## **plt.legend()**
The Legend class is a container of legend handles and legend texts.

### **loc**
The location of the legend.

(default: 'best') ('best' for axes, 'upper right' for figures)

The strings 'upper left', 'upper right', 'lower left', 'lower right' place the legend at the corresponding corner of the axes/figure.

The strings 'upper center', 'lower center', 'center left', 'center right' place the legend at the center of the corresponding edge of the axes/figure.

The string 'center' places the legend at the center of the axes/figure.

```python
x=[150,160,165,185]
y=[70,80,90,100]
plt.plot(x,y)
plt.title("Title of graph",color='b')
#x -axis label 
plt.xlabel("x-axis labels",c='red')
#y-axis label
plt.ylabel("y-axis labels",color='m')

# Add legend in lower right part of figure 
plt.legend(labels=['ca','w'],loc='lower right')
```

## **plt.figure()** 

###  Parameter  - figsize()
You can specify the size of figure using figure() and passinG tuple consisting of rows and columns to argument figsize.The values of length are in inches

```python
x=[150,160,165,185]
y=[70,80,90,100]
plt.figure(figsize = (9, 7)) 
plt.plot(x, y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("Title of Line Chart")
plt.show()
```

## **plt.xticks()**
Get or set the current tick locations and labels of the x-axis.

### **Parameter**

### **ticks** 
ticksarray-like, optional
The list of xtick locations. Passing an empty list removes all xticks.

### **labels**
labelsarray-like, optional
The labels to place at the given ticks locations. This argument can only be passed if ticks is passed as well.

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [8, 4, 6, 4, 9, 4, 2]
plt.plot(x, y, 'b') # b specifies blue
plt.xlabel("x value") # sets the label for x-axis of the plot
plt.ylabel("y value") # set the label for y-axis of the plot
plt.title("Plot x y") # sets the title of plot
plt.xticks([2, 4, 6]) # sets ticks for x-axis
plt.show()
```

## **Annotations **

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [8, 4, 6, 4, 9, 4, 2]
plt.plot(x, y, 'b') # b specifies blue
plt.xlabel("x value") # sets the label for x-axis of the plot
plt.ylabel("y value") # set the label for y-axis of the plot
plt.title("Plot x y") # sets the title of plot
plt.text(5, 9, "Max y") # (5, 9) is the data point to annotate
plt.title("Plot x y") # sets the title of plot
plt.show()
```

## **Format Strings**

A format string consists of a part for color, marker and line:

fmt = '[marker][linestyle][color]'
Each of them is optional. If not provided, the value from the style cycle is used. Exception: If line is given, but no marker, the data will be a line without markers.

Other combinations such as [color][marker][line] are also supported, but note that their parsing may be ambiguous.

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [8, 4, 6, 4, 9, 4, 2]
plt.figure(figsize = (9, 7)) 
plt.plot(x, y,'x-.r')
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("Plot with markers")
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [8, 4, 6, 4, 9, 4, 2]
plt.figure(figsize = (9, 7)) 
plt.plot(x, y,marker='o',ls='--',c='k')
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("Plot with markers")
plt.show()
```

## **line width ,markersize**

Note : while using in plt pass fmt just after passing x,y and then after passing fmt string pass line width or markersize

### **linewidth**
setting width for line 
linewidth=10

### **markersize**
setting marker size 
markersize=8

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [8, 4, 6, 4, 9, 4, 2]
plt.figure(figsize = (9, 7)) 
plt.plot(x, y,'X-.r',markersize=12,linewidth=2)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("Plot with markers")
plt.show()
```

# **Multiple  line charts or Stacked Line charts**
*Note* : without using legend() , we can have it by using label .It is only for mutliple line charts or stacked line chart.

```python
day = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
baby_one = [15,13,17,9,11,19,13]
baby_two = [12,10,18,14,16,20,7]
line_chart1 = plt.plot(day, baby_one, color='Blue',ls= '--')
line_chart2 = plt.plot(day, baby_two, color='Red',ls= '-.')
plt.xlabel('Day',color="red")
plt.ylabel('Birth Rate',color="yellow")
plt.title('Arctic hospital last week birth rate')
plt.legend(['1-7 July 2019', '8-14 July 2019'], loc=3)
```

```python
days = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
company1 = [50, 40, 23, 34, 20, 10, 50] # sales of company1
company2 = [30, 40, 30, 45, 20, 45, 30] # sales of company 2
plt.figure(figsize = (10, 8))
plt.plot(days, company1,'*r-.', label = "Company 1") # line chart company 1
plt.plot(days, company2, label = "Company 2") # line chart company 2
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend() # using legend to differentiate the graphs of companies
plt.title("Sales data of company 1 and company 2 for this week")
plt.show()
```

# ** 3d -Line Plot **

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 8))
ax = plt.axes(projection="3d")
plt.title("3D line plot")
# Data for a three-dimensional line
xline = [1,2,3,4,5,6,7,8]
yline = [1,2,3,4,5,6,7,8]
zline = [20, 30, 20, 40, 10, 50, 20, 30]
ax.plot3D(xline, yline, zline, 'cyan')
ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_zlabel("z values")
plt.show()
```
